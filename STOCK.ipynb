```python
# Create a new dataframe with only the 'Close column 
import numpy as np
Ndata = NVDA.filter(['close'])
# Convert the dataframe to a numpy array
Ndataset = Ndata.values
# Get the number of rows to train the model on
training_Ndata_len = int(np.ceil( len(Ndataset) * .95 ))

training_Ndata_len
```

```python
# Scale the data
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0,1))
scaled_Ndata = scaler.fit_transform(Ndataset)

scaled_Ndata
```

```python
# Create the training data set 
# Create the scaled training data set
train_Ndata = scaled_Ndata[0:int(training_Ndata_len), :]
# Split the data into x_train and y_train data sets
x_Ntrain = []
y_Ntrain = []
#滑动窗口选取60的原因是为了跟5%测试集对上
for i in range(60, len(train_Ndata)):
    x_Ntrain.append(train_Ndata[i-60:i, 0])
    y_Ntrain.append(train_Ndata[i, 0])
    if i<= 61:
        print(x_Ntrain)
        print(y_Ntrain)
        print()          #这里if语句是为了观察两组数据
        
# Convert the x_train and y_train to numpy arrays 
x_Ntrain, y_Ntrain = np.array(x_Ntrain), np.array(y_Ntrain)

# Reshape the data
x_Ntrain = np.reshape(x_Ntrain, (x_Ntrain.shape[0], x_Ntrain.shape[1], 1))
# x_train.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM

# Build the LSTM model
model = Sequential()
model.add(LSTM(128, return_sequences=True, input_shape= (x_Ntrain.shape[1], 1)))
model.add(LSTM(64, return_sequences=False))
#model.add(LSTM(32,return_sequences=False))
model.add(Dense(50))
model.add(Dense(25))
model.add(Dense(1))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(x_Ntrain, y_Ntrain, batch_size=1, epochs=1)
```

```python
# Create the testing data set
test_Ndata = scaled_Ndata[training_Ndata_len - 60: , :]#training_Ndata_len - 60 表示从倒数第60个元素开始选择，直到数组的末尾。
# Create the data sets x_test and y_test
x_Ntest = []
y_Ntest = Ndataset[training_Ndata_len:, :]
for i in range(60, len(test_Ndata)):
    x_Ntest.append(test_Ndata[i-60:i, 0])
    
# Convert the data to a numpy array
x_Ntest = np.array(x_Ntest)
# Reshape the data
x_Ntest = np.reshape(x_Ntest, (x_Ntest.shape[0], x_Ntest.shape[1], 1 ))

# Get the models predicted price values 
predictions = model.predict(x_Ntest)
predictions = scaler.inverse_transform(predictions)

# Get the root mean squared error (RMSE)
rmse = np.sqrt(np.mean(((predictions - y_Ntest) ** 2)))
rmse
```

```python
mae = np.mean(np.abs(predictions - y_Ntest))
mae                     
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_Ntest, predictions)
r2
```

```python
dar = np.mean((np.sign(predictions - y_Ntest) + 1) / 2)
dar
```

```python
# Plot the data
Ntrain = Ndata[:training_Ndata_len]
Nvalid = Ndata[training_Ndata_len:]
Nvalid['Predictions'] = predictions
# Visualize the data
plt.figure(figsize=(15,10))
plt.title('Model')
plt.xlabel('Date', fontsize=18)
plt.ylabel('Close Price NVDA ($)', fontsize=18)
plt.plot(Ntrain['close'])
plt.plot(Nvalid[['close', 'Predictions']])
plt.legend(['Train', 'Val', 'Predictions'], loc='lower right')
plt.show()
Nvalid
```
